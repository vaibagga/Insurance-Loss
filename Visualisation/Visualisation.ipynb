```python
import numpy as np
import pandas as pd 
import scipy
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('../Data/Compressed.csv')
train.describe()
```

```python
def plotHistData(train):
    i = 0
    for col in train.columns.values:
        if train[col].dtype == 'int32' or train[col].dtype == 'float32':
            plt.figure(i)
            i+=1
            plt.xlabel(col)
            plt.ylabel('Frequency')
            plt.hist(train[col])
```

```python
plotHistData(train)
```

```python
## scatterlots
def plotScatterData(train):
    i = 0
    for col1 in train.columns.values:
        for col2 in train.columns.values:
            if (train[col1].dtype == 'int32' or train[col1].dtype == 'float32') and (train[col2].dtype == 'int32' or train[col2].dtype == 'float32') and col1 != col2: 
                plt.figure(i)
                i+=1
                plt.xlabel(col1)
                plt.ylabel(col2)
                plt.scatter(train[col1], train[col2])
```

```python
plotScatterData(train)
```

```python
corr = train.corr()
sns.heatmap(corr)
```
